* Run basic calculations to obtain number of unique items, average price, etc.

```python
items = purchase_data['Item ID'].nunique()
price_avg = purchase_data['Price'].mean()
purchases = purchase_data['Purchase ID'].count()
revenue = purchase_data['Price'].sum()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender = purchase_data.groupby(by='Gender',sort=False)['SN'].nunique()
gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_df = purchase_data.groupby(by='Gender')['Purchase ID'].count()
gender_df = pd.DataFrame(gender_df)
del gender_df.index.name
gender_df
```

```python
gender_df['APP'] = purchase_data.groupby(by='Gender')['Price'].mean()
gender_df['TPV'] = purchase_data.groupby(by='Gender')['Price'].sum()
gender_df
```

```python
gender = purchase_data.groupby(by='Gender')['SN'].nunique()
gender = pd.DataFrame(gender)
del gender.index.name
# gender.columns = ['Total Count']
# gender['SN'] = gender_analysis.TPV/gender.SN.sum()
gender['TPV'] = purchase_data.groupby(by='Gender')['Price'].sum()
gender['ATPPP'] = gender.TPV/gender.SN
gender
```

```python
gender_analysis = gender_df.merge(gender,right_index=True,left_index=True)
gender_analysis = gender_analysis[['Purchase ID','APP','TPV_x','ATPPP']]

gender_analysis['APP'] = gender_analysis['APP'].astype(float).map("${:,.2f}".format)
gender_analysis['TPV_x'] = gender_analysis['TPV_x'].astype(float).map("${:,.2f}".format)
gender_analysis['ATPPP'] = gender_analysis['ATPPP'].astype(float).map("${:,.2f}".format)

gender_analysis.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value','Avg Total Purchase per Person']
gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,110]
group_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Range'] = pd.cut(purchase_data['Age'], bins, labels=group_names)
purchase_data.head()
```

```python
age_df = purchase_data.groupby(by='Age Range')['SN'].nunique()
age_df = pd.DataFrame(age_df)
del age_df.index.name
age_df
```

```python
age_df['Percentage of Players'] = age_df.SN/age_df.SN.sum()*100
age_df['Percentage of Players'] = age_df['Percentage of Players'].astype(float).map("{:,.2f}".format)
age_df.columns = ['Total Count','Percentage of Players']
age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_cnt = pd.DataFrame(purchase_data['SN'].value_counts())
avg_price = pd.DataFrame(purchase_data.groupby(by='SN')['Price'].sum())
total_purchase = pd.DataFrame(purchase_data.groupby(by='SN')['Price'].sum())
```

```python
top_spend = purchase_cnt.merge(avg_price,left_index=True,right_index=True)
top_spend['Price'] = top_spend.Price/top_spend.SN
top_spend.head()
```

```python
top_spend = top_spend.merge(total_purchase,left_index=True,right_index=True)
top_spend.head()
```

```python
top_spend = top_spend.nlargest(5, 'Price_y')
```

```python
top_spend.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value']

top_spend['Average Purchase Price'] = top_spend['Average Purchase Price'].astype(float).map("${:,.2f}".format)
top_spend['Total Purchase Value'] = top_spend['Total Purchase Value'].astype(float).map("${:,.2f}".format)

top_spend
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_nm = purchase_data.set_index(['Item ID','Item Name'])
item_nm = item_nm[~item_nm.index.duplicated(keep='first')]
item_nm = item_nm[['Price']]
item_nm.head()
```

```python
item_pc = pd.DataFrame(purchase_data.groupby(by='Item ID')['Purchase ID'].count())
item_pc.head()
```

```python
item_pop = item_pc.merge(item_nm,left_index=True,right_index=True)
item_pop.columns = ['Purchase Count','Item Price']
item_pop['Total Purchase Value'] = item_pop['Purchase Count']*item_pop['Item Price']

item_pop['Item Price'] = item_pop['Item Price'].astype(float).map("${:,.2f}".format)
item_pop['Total Purchase Value'] = item_pop['Total Purchase Value'].astype(float).map("${:,.2f}".format)

item_pop.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_prof = item_pc.merge(item_nm,left_index=True,right_index=True)
item_prof.columns = ['Purchase Count','Item Price']
item_prof['Total Purchase Value'] = item_prof['Purchase Count']*item_prof['Item Price']

item_prof = item_prof.sort_values(by=['Total Purchase Value'], ascending=False)

item_prof['Item Price'] = item_prof['Item Price'].astype(float).map("${:,.2f}".format)
item_prof['Total Purchase Value'] = item_prof['Total Purchase Value'].astype(float).map("${:,.2f}".format)

item_prof.head()
```
